# Movie Reviews

```python
import pandas as pd

data = pd.read_pickle("reviews")

data.head()
```

The dataset is made up of positive and negative movie reviews.

## Preprocessing

üëá Remove punctuation and lower case the text.

```python
import string
```

```python
def remove_punct_low(reviews):
    text = "".join([word for word in reviews if word not in string.punctuation])
    return text.lower()


data['reviews'] = data['reviews'].apply(remove_punct_low)
data['reviews']
```

## Bag-of-Words modelling

üëá Using `cross_validate`, score a Multinomial Naive Bayes model trained on a Bag-of-Word representation of the texts.

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data['reviews'])

X_bow = X.toarray()

pd.DataFrame(X_bow)
```

```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score

nb_model = MultinomialNB()
cv_scores = cross_val_score(nb_model, X_bow, data['target'], cv=5)
print("Accuracy: " , cv_scores.mean())
```

## N-gram modelling

üëá Using `cross_validate`, score a Multinomial Naive Bayes model trained on a 2-gram Bag-of-Word representation of the texts.

```python
# Bigram = 2-gram : ensemble de 2 mots
```

```python
vectorizer = CountVectorizer(ngram_range = (2,2))
Y = vectorizer.fit_transform(data['reviews'])

Y.toarray()
```

```python
nb_model = MultinomialNB()
cv_scores = cross_val_score(nb_model, Y, data['target'], cv=10)
print("Accuracy: " , cv_scores.mean())
```
